Using the below code snippet as an example of how to deploy the [cross_validation](http://scikit-learn.org/stable/modules/cross_validation.html) module, reload the dataset that was used in the last exercise (look it up) and apply Kfold cross validation to it. What observations do you have?

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import cross_validation

# Load data from csv file
df = pd.read_csv('data/balance.csv')
X = df[['Income', 'Limit']].values
y = df['Balance'].values


kf = cross_validation.KFold(X.shape[0], n_folds=5, shuffle=True, random_state=10)
results = []
for train_index, test_index in kf:
    regr = LinearRegression()
    regr.fit(X[train_index], y[train_index])
    results.append(regr.score(X[test_index], y[test_index]))
print(results)
print("average score:", np.mean(results))

```
