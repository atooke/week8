```python
from sklearn.datasets import load_iris

iris_data = load_iris()
print(iris_data.data.shape)
print(iris_data.DESCR)
```

```python
iris_data.feature_names
```

```python
columns = [name[:-5].replace(' ','_') for name in iris_data.feature_names]
columns
```

```python
iris_data.data
```

```python
import pandas as pd
import numpy as np

iris_df = pd.DataFrame(iris_data.data, columns=columns)
iris_df
```

```python
iris_df['label'] = iris_data.target_names[iris_data.target]
iris_df['target'] = iris_data.target
```

```python
iris_df.head()
```

Let's take a quick look at our data by plotting some of it. Here, we're going to use a scatter plot of sepal_length vs sepal_width and color by the label. You'll easily be able to see that one of the classes is very different from the other two.

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(1, 1, 1)
for target, color in zip(iris_df.target.unique(), ['r', 'g', 'b']):
    sub_df = iris_df.query('target == @target')
    ax.scatter(x=sub_df.sepal_length.values, y=sub_df.sepal_width.values, 
               color=color, label=iris_data.target_names[target], s=30)
ax.legend(loc='best')
ax.set_xlabel('Sepal Length', size=15)
ax.set_ylabel('Sepal Width', size=15)
```

It seems that a model might have a good chance at being able to distinguish between classes based solely on sepal length and width (adding in petal length and width is only going to make that better). We almost always want to do more EDA than we've done here, but that isn't the focus of tonight's lecture. To try and predict the class of an unknown flower with known measurements, we are going to use logistic regression.

```python
X = iris_df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = iris_df.label
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression()
logistic_model.fit(X, y)
```

Now, we can see what our model would predict if it saw data with an unknown label. What would we predict for a flower with sepal length = 4.5, sepal_width = 3.3, petal_length = 1.6 and petal_width = 0.2? To find out, we can use the predict method on our newly fitted model! We'll see that it thinks it's a "Setosa". It turns out that our fitted model will also accept a 2-dimensional array and hand back preditions for each row in this array.

```python
print('Model guess for single point: {}'.format(logistic_model.predict(np.array([[4.5, 3.3, 1.6, 0.2]]))[0]))
logistic_model.predict(np.array([[7.2, 2.8, 6.6, 2], [6.2, 2.5, 3.6, 2], [4.7, 3.6, 1.9, .1]]))
```
